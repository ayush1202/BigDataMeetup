```python
'''Boxplot: Adding a layer of individual points on top of it through 
Seaborn's striplot. We'll use jitter=True so that all the points don't fall in single vertical lines
above the species. Saving the resulting axes as ax each time causes the resulting plot to be shown
on top of the previous axes'''
ax = sns.boxplot(x="Yr", y="Jan_oil", data=data_viz)
ax = sns.stripplot(x="Yr", y="Jan_oil", data=data_viz, jitter=True, edgecolor='gray')
```

```python
'''A violin plot combines the benefits of the previous two plots and simplifies them.
Denser regions of the data are fatter, and sparser thiner in a violin plot'''
sns.violinplot(x="Yr", y="Jan_oil", data=data_viz, size=6)
```

```python
'''A final seaborn plot useful for looking at univariate relations is the kdeplot, 
Creates and visualizes a kernel density estimate of the underlying feature'''

sns.FacetGrid(data_viz, hue="County", size=6) \
   .map(sns.kdeplot, "Yr") \
   .add_legend()
```

```python
# pairplot, which shows the bivariate relation
# between each pair of features
sns.pairplot(data_viz, hue="Yr", size=3)
```

```python
# The diagonal elements in a pairplot show the histogram by default
# We can update these elements to show other things, such as a kde
sns.pairplot(data_viz, hue="Yr", size=3, diag_kind="kde")
```

```python
# Importing the additional libraries
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
# Running a SQLite Query to look into just the oil production - Top 100 based on Jan production
# To be noted here that the columns ordered based on the query
data_expl= pd.read_sql_query('''
SELECT Apino, Yr, Wellname, Company, County, Formation, Jan_oil, Feb_oil, Mar_oil, Apr_oil, May_oil, 
Jun_oil, Jul_oil, Aug_oil, Sep_oil, Oct_oil, Nov_oil, Dec_oil
FROM Production
ORDER BY Jan_oil DESC
LIMIT 100;''', conn)

data_expl
```

```python
data_expl.describe() # very basic descriptive statistics for entire dataset
data_expl['Jan_oil'].describe() # descriptive statistics for a particular column
```

```python
#plotting the histogram
sns.distplot(data_expl['Jan_oil']);

# Observations:
# 1. Deviations from the Normal Distribution
# 2. Positive Skewness
# 3. Shows peakedness
```

```python
#skewness and kurtosis
print("Skewness: %f" % data_expl['Jan_oil'].skew())
print("Kurtosis: %f" % data_expl['Jan_oil'].kurt())
```

```python
# Note: If we need to plot two different datasets on the same plot, use common plotname
fig = plt.scatter(data_expl.Jan_oil, data_expl.Feb_oil)
fig = plt.scatter(data_expl.Mar_oil, data_expl.Apr_oil)
```

```python
sns.boxplot(x="Yr", y="Jan_oil", data=data_expl)
```

```python
# Running a SQLite Query to look into just the gas production
# To be noted here that the columns ordered based on the query
data_expl2= pd.read_sql_query('''
SELECT 
Apino, Yr, Wellname, Company, County, Formation, Jan_oil, Feb_oil, Mar_oil, Apr_oil, May_oil, Jun_oil, Jul_oil, 
Aug_oil, Sep_oil, Oct_oil, Nov_oil, Dec_oil, (Jan_oil + Feb_oil + Mar_oil + Apr_oil + May_oil + Jun_oil + 
Jul_oil + Aug_oil + Sep_oil + Oct_oil + Nov_oil + Dec_oil) AS Sum

FROM Production
ORDER BY Jan_oil DESC
LIMIT 100;''', conn)

data_expl2
```

```python
# Correlation Matrix

corr_matrix = data_expl2.corr()
f, ax = plt.subplots(figsize = (6, 6))
sns.heatmap(corr_matrix, vmax=0.8, square=True)
```

```python
# Checking for the following statistical parameters
# 1. Normality - checking for the normal distribution
# 2. Homoscedasticity - assumption that dependent variables exhibit equal levels of variance across the range of predictor variables
# 3. Linearity - Good idea to check in case any data transformation is required
# 4. Absence of correlated errors

sns.distplot(data_expl2['Sum'],fit=norm)
fig = plt.figure()
res = stats.probplot(data_expl2['Sum'], plot=plt)

# Histogram (for kurtosis and skewness) and Normal Probability Plot (data distribution should follow the diagonal)
```

```python
# applying log transformation
# in case of positive skewness, log transformations usually works well
data_expl2['Sum'] = np.log(data_expl2['Sum'])
```

```python
sns.distplot(data_expl2['Sum'],fit=norm)
fig = plt.figure()
res = stats.probplot(data_expl2['Sum'], plot=plt)
```

```python
# number of missing values in each column, here we are using the entire dataset - "data" dataframe
missing_df = data.isnull().sum(axis=0).reset_index()
missing_df.columns = ['column_name', 'missing_count']
missing_df = missing_df.ix[missing_df['missing_count']>0]
missing_df = missing_df.sort_values(by='missing_count')

ind = np.arange(missing_df.shape[0])
width = 0.9
fig, ax = plt.subplots(figsize=(5,5))
rects = ax.barh(ind, missing_df.missing_count.values, color='blue')
ax.set_yticks(ind)
ax.set_yticklabels(missing_df.column_name.values, rotation='horizontal')
ax.set_xlabel("Count of missing values")
ax.set_title("Number of missing values in each column")
plt.show()
```
